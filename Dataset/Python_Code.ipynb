```python
import pandas as pd 
import numpy as np 
df = pd.read_csv("clean_data.csv")
```

```python
channel_performance = (
    df
    .groupby('MARKETING_CHANNEL')
    .agg(
        total_revenue=('GBP_PRICE', 'sum'),
        avg_order_value=('GBP_PRICE', 'mean'),
        total_orders=('ORDER_ID', 'nunique')
    )
    .sort_values('total_revenue', ascending=False).round(2)
)
print(channel_performance)
```

```python
orders_per_user = df.groupby('USER_ID')['ORDER_ID'].nunique()
orders_per_user.value_counts()

median_order_value = df['GBP_PRICE'].median()

df['order_value_segment'] = df['GBP_PRICE'].apply(
    lambda x: 'High' if x > median_order_value else 'Low'
)

order_value_summary = (
    df.groupby('order_value_segment')
      .agg(
          order_count=('ORDER_ID', 'count'),
          total_revenue=('GBP_PRICE', 'sum')
      )
)

print(order_value_summary)
```
